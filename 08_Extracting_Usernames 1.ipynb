### Extracting Usernames - Extract the usernames from the email addresses present in the text


#### 1. Using Regular Expressions

```python
import re
```

```python
def extract_usernames(text):
    # Define a regular expression pattern for extracting email usernames
    pattern = re.compile(r'(\S+)@')

    # Use the findall method to extract all matches
    usernames = pattern.findall(text)

    return usernames
```

```python
# Example usage:
text_with_emails = "Here are some email addresses: user1@example.com, user2@example.com, user3@example.com"
result = extract_usernames(text_with_emails)
```

```python
print("Extracted Usernames:", result)
```

#### 2. Using spaCy

```python
import spacy
!python -m spacy download en_core_web_md
nlp = spacy.load('en_core_web_sm')

def extract_usernames_spacy(text):
    doc = nlp(text)
    usernames = [token.text.split('@')[0] for token in doc if '@' in token.text]

    return usernames

# Example usage:
text_with_emails = "Here are some email addresses: user1@example.com, user2@example.com, user3@example.com"
result_spacy = extract_usernames_spacy(text_with_emails)

print("Extracted Usernames (spaCy):", result_spacy)
```
